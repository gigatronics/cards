```python
parts = []

for url in urls:
    r = requests.get(url)
    r.raise_for_status()
    parts.append(r.text)

combined_html = "\n\n<!-- PAGE BREAK -->\n\n".join(parts)

with open("taw_test.html", "w", encoding="utf-8") as f:
    f.write(combined_html)
```

```python
## no don't clean! 
# 
# 
def clean_soup(soup):
    # Remove scripts, styles, ads
    for tag in soup(['script', 'style', 'nav', 'footer', 'div']):
        if tag.get('class') and ('ad' in tag.get('class') or 'sidebar' in tag.get('class')):
            tag.decompose()
    
    # Replace breaks with newlines
    for br in soup.find_all("br"):
        br.replace_with("\n")
    
    return soup.get_text(separator="\n", strip=True)

plain_soup = clean_soup(soup)
plain_soup# .find_all(id = "disclaimer-token-UDINQEN3")
```
